```python
from martingale import MartinGale
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('default')
sns.set()
sns.set_style('whitegrid')
sns.set_palette('gray')
```

```python
funds = 2500 # 所持金
game_type = 'roulette' # プレイするゲーム
initial_bet = 10 # martingaleの最初の掛け金
simulation_num = 720# 賭ける回数

funds_list = []
for i in range(10000):
    marting_gale = MartinGale(funds, game_type, initial_bet, simulation_num)
    funds_list.append(marting_gale.simulate())
result_list = [log[-1] for log in funds_list]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(result_list)
ax.set_title('histogram of resulting cash')
ax.set_xlabel('funds')
ax.set_ylabel('freq')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(funds_list[0])
ax.set_title('sampled transition of cash')
ax.set_xlabel('number of plays')
ax.set_ylabel('cash')
```

```python
print(len([res for res in result_list if res >= funds]) / 10000)
print(len([res for res in result_list if res == 0]) / 10000)
```
